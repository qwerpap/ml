#### 4. Сравните баллы у студентов разных полов. Используя agg() выведите минимальное, максимальное и медианное значение


```python
#Шаг 1: Группировка данных по полу и применение функции agg для расчета минимального, максимального и медианного значений
# Метод agg() позволяет применять несколько агрегирующих функций одновременно
result = df.groupby('gender').agg({
   'math score': ['min', 'max', 'median'],
   'reading score': ['min', 'max', 'median'],
   'writing score': ['min', 'max', 'median']
})

# Шаг 2: Вывод результата
print("Сравнение баллов у студентов разных полов (минимальное, максимальное и медианное значение):")
print(result)
```

#### 8. Постройте plot.pie, показывающий, сколько людей сдали\не сдали экзамен по математике.

Сдавшим считается человек, набравший 40 баллов.

#### 9. Постройте plot.pie, показывающий распределение студентов по оценкам

Оценки студентов выставляются по шкале:<br>
0  - 40 marks : grade E<br>
41 - 60 marks : grade D<br>
60 - 70 marks : grade C<br>
70 - 80 marks : grade B<br>
80 - 90 marks : grade A<br>
90 - 100 marks : grade O<br>

Для этого посчитайте сумму результатов за 3 экзамена и найдите среднее. Оценка выставляется по среднему значению. Если студент не сдал математику(даже если средний балл выше 40), он получает Е

```python
# Шаг 1: Рассчитываем средний балл по трем предметам
df['average_score'] = (df['math score'] + df['reading score'] + df['writing score']) / 3

# Шаг 2: Определяем оценку для каждого студента по средней оценке и проверке на сдачу математики
def determine_grade(row):
    # Если студент не сдал математику (менее 40 баллов), ставим E
    if row['math score'] < 40:
        return 'E'
    # Оценка по среднему баллу
    elif row['average_score'] <= 40:
        return 'E'
    elif 41 <= row['average_score'] <= 60:
        return 'D'
    elif 61 <= row['average_score'] <= 70:
        return 'C'
    elif 71 <= row['average_score'] <= 80:
        return 'B'
    elif 81 <= row['average_score'] <= 90:
        return 'A'
    else:
        return 'O'

# Применяем функцию ко всем строкам для расчета оценки
df['grade'] = df.apply(determine_grade, axis=1)

# Шаг 3: Подсчитываем количество студентов для каждой оценки
grade_counts = df['grade'].value_counts()

# Шаг 4: Построение круговой диаграммы
plt.figure(figsize=(6, 6))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'orange', 'red', 'purple', 'gold'])
plt.title('Распределение студентов по оценкам')
plt.show()
```

#### 10. Постройте countplot, показывающий зависимость между итоговой оценкой студентов и его полом. Студенты какого пола получили больше оценок О, А, В

```python
# Шаг 1: Рассчитываем средний балл по трем предметам
df['average_score'] = (df['math score'] + df['reading score'] + df['writing score']) / 3

# Шаг 2: Определяем итоговую оценку для каждого студента по средней оценке и проверке на сдачу математики
def determine_grade(row):
    # Если студент не сдал математику (менее 40 баллов), ставим E
    if row['math score'] < 40:
        return 'E'
    # Оценка по среднему баллу
    elif row['average_score'] <= 40:
        return 'E'
    elif 41 <= row['average_score'] <= 60:
        return 'D'
    elif 61 <= row['average_score'] <= 70:
        return 'C'
    elif 71 <= row['average_score'] <= 80:
        return 'B'
    elif 81 <= row['average_score'] <= 90:
        return 'A'
    else:
        return 'O'

# Применяем функцию ко всем строкам для расчета итоговой оценки
df['final_grade'] = df.apply(determine_grade, axis=1)

# Шаг 3: Построение countplot для визуализации распределения оценок в зависимости от пола
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='final_grade', hue='gender', palette='Set2')

# Настройка графика
plt.title('Распределение оценок по полу', fontsize=16)
plt.xlabel('Итоговая оценка', fontsize=12)
plt.ylabel('Количество студентов', fontsize=12)
plt.legend(title='Пол студента')

# Отображение графика
plt.tight_layout()
plt.show()
```
