10) Создать матрицу случайных чисел размера 10 на 10 и отнять среднее по строкам из каждого элемента строки

```python
import numpy as np  # Импортируем библиотеку numpy для работы с массивами и матрицами

# Шаг 1: Создаем матрицу случайных чисел размера 10x10
matrix = np.random.rand(10, 10)

# Выводим исходную матрицу
print("Исходная матрица:\n", matrix)

# Шаг 2: Вычисляем среднее значение по каждой строке
row_means = matrix.mean(axis=1, keepdims=True)

# Выводим средние значения для каждой строки
print("Средние значения по строкам:\n", row_means)

# Шаг 3: Отнимаем среднее значение каждой строки из элементов этой строки
matrix_centered = matrix - row_means

# Выводим полученную матрицу
print("Матрица после вычитания среднего по строкам:\n", matrix_centered)
```

11) Создать матрицу случайных чисел размера 10 на 10 и отнять медиану по столбцам из каждого элемента столбца  

```python
import numpy as np  # Импортируем библиотеку numpy для работы с массивами и матрицами

# Шаг 1: Создаем матрицу случайных чисел размера 10x10
matrix = np.random.rand(10, 10)

# Выводим исходную матрицу
print("Исходная матрица:\n", matrix)

# Шаг 2: Вычисляем медиану по каждому столбцу
column_medians = np.median(matrix, axis=0, keepdims=True)

# Выводим медианы по столбцам
print("Медианы по столбцам:\n", column_medians)

# Шаг 3: Отнимаем медиану каждого столбца из элементов этого столбца
matrix_centered = matrix - column_medians

# Выводим матрицу после вычитания медиан по столбцам
print("Матрица после вычитания медиан по столбцам:\n", matrix_centered)
```

12) Создать матрицу случайных чисел 5x5. Добавить к ней столбецы максимумов и минимумов по строкам. Получится матрица 5x7.  

```python
import numpy as np  # Импортируем библиотеку numpy для работы с массивами и матрицами

# Шаг 1: Создаем матрицу случайных чисел размером 5x5
matrix = np.random.rand(5, 5)

# Выводим исходную матрицу
print("Исходная матрица 5x5:\n", matrix)

# Шаг 2: Вычисляем максимальные значения по каждой строке
max_in_rows = np.max(matrix, axis=1, keepdims=True)

# Выводим максимальные значения по строкам
print("Максимумы по строкам:\n", max_in_rows)

# Шаг 3: Вычисляем минимальные значения по каждой строке
min_in_rows = np.min(matrix, axis=1, keepdims=True)

# Выводим минимальные значения по строкам
print("Минимумы по строкам:\n", min_in_rows)

# Шаг 4: Добавляем столбцы максимумов и минимумов к исходной матрице
matrix_with_max_min = np.hstack((matrix, max_in_rows, min_in_rows))

# Выводим итоговую матрицу 5x7
print("Матрица 5x7 (с максимумами и минимумами):\n", matrix_with_max_min)
```

13) Вектор чисел от 1 до 51 преобразовать в вектор нечетных чисел и в вектор четных чисел, поменять их размерности, чтобы получились две матрица. Найти произведение этих матриц. Найти обратную матрицу эётого произведения.  

```python
import numpy as np  # Импортируем библиотеку numpy для работы с массивами и матрицами

# Шаг 1: Создаем вектор чисел от 1 до 51
vector = np.arange(1, 52)

# Шаг 2: Преобразуем вектор в вектор нечетных чисел и вектор четных чисел
odd_numbers = vector[vector % 2 != 0]  # Нечетные числа
even_numbers = vector[vector % 2 == 0]  # Четные числа

# Выводим векторы нечетных и четных чисел
print("Вектор нечетных чисел:\n", odd_numbers)
print("Вектор четных чисел:\n", even_numbers)

# Шаг 3: Меняем размерности векторов, чтобы получились матрицы
odd_matrix = odd_numbers.reshape(-1, 1)  # Преобразуем в матрицу с одной колонкой
even_matrix = even_numbers.reshape(1, -1)  # Преобразуем в матрицу с одной строкой

# Выводим полученные матрицы
print("Матрица из нечетных чисел:\n", odd_matrix)
print("Матрица из четных чисел:\n", even_matrix)

# Шаг 4: Находим произведение матриц
product_matrix = np.dot(odd_matrix, even_matrix)

# Выводим произведение матриц
print("Произведение матриц:\n", product_matrix)

# Шаг 5: Проверяем, является ли произведение квадратной матрицей и обратимо ли оно
if product_matrix.shape[0] == product_matrix.shape[1]:
    # Шаг 6: Находим обратную матрицу
    inverse_matrix = np.linalg.inv(product_matrix)
    # Выводим обратную матрицу
    print("Обратная матрица:\n", inverse_matrix)
else:
    print("Произведение не является квадратной матрицей, обратная матрица не может быть вычислена.")
```

14) Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами.

```python
import numpy as np  # Импортируем библиотеку numpy для работы с массивами и матрицами

# Шаг 1: Создаем пустую матрицу размером 8x8
chessboard = np.zeros((8, 8), dtype=int)

# Шаг 2: Заполняем матрицу в шахматном порядке единицами
chessboard[1::2, ::2] = 1  # Заполняем нечетные строки (1, 3, 5, 7) на четных позициях (0, 2, 4, 6)
chessboard[::2, 1::2] = 1  # Заполняем четные строки (0, 2, 4, 6) на нечетных позициях (1, 3, 5, 7)

# Выводим полученную матрицу
print("Матрица 8x8, заполненная в шахматном порядке нулями и единицами:\n", chessboard)
```
